Задание:
1. Сгенерировать две выборки Х1 и Х2 объема N = 100 из нормального
распределения с параметрами, удовлетворяющими следующим ситуациям:

a. Математические ожидания и дисперсии равны;

b. Математические ожидания существенно различны, дисперсии
равны;

c. Математические ожидания равны, дисперсии существенно
различны;

d. Математические ожидания и дисперсии существенно различны.

2. Для каждой ситуации:

a. Проверить гипотезы о нормальности с помощью критерия
Шапиро-Уилка, гипотезы о равенстве средних и дисперсий;

b. На одном графике построить гистограммы Х1 и Х2, оценки
плотностей для обеих выборок, вертикальными линиями отметить
средние значения;

c. На другом графике построить коробки с усами для Х1 и Х2.

```r
# 1 - a
mu1_a = mu2_a = 2.25
sigma1_a = sigma2_a = 1.75
N = 100
X1 = rnorm(N, mu1_a, sigma1_a)
X2 = rnorm(N, mu2_a, sigma2_a)

# 2 - a
shapiro.test(X1)
shapiro.test(X2)

t.test(X1, X2, var.equal = TRUE)
var.test(X1, X2)

# 2 - b
library(ggplot2)

df <- data.frame(
  value = c(X1, X2),
  group = rep(c("X1", "X2"), each = N)
)

ggplot(df, aes(x = value, fill = group)) +
  geom_histogram(aes(y = ..density..), position = "identity", alpha = 0.5, bins = 20) +
  geom_density(alpha = 0.5) +
  geom_vline(aes(xintercept = mean(X1)), color = "blue", linetype = "dashed", size = 1) +
  geom_vline(aes(xintercept = mean(X2)), color = "red", linetype = "dashed", size = 1) +
  labs(title = "Гистограммы и оценки плотности X1 и X2", x = "Значение", y = "Плотность") +
  theme_minimal()

# 2 - c
ggplot(df, aes(x = group, y = value, fill = group)) +
  geom_boxplot() +
  labs(title = "Boxplot для X1 и X2", x = "Группа", y = "Значение") +
  theme_minimal()
```
